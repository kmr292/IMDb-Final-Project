```python
import pandas as pd
import numpy as np
```

```python
# Reading csv file
IMDB_movies = pd.read_csv('IMDB_movies_and_ratings.csv')
# Dropping unneeded columns
IMDB_movies = IMDB_movies.drop(columns = ['Unnamed: 0', 'original_title','non_us_voters_rating', 'non_us_voters_votes', 'imdb_title_id', 'date_published', 'top1000_voters_rating', 
                                          'top1000_voters_votes', 'males_0age_avg_vote', 'males_0age_votes',
                                          'males_18age_avg_vote', 'males_18age_votes', 'males_30age_avg_vote', 'males_30age_votes', 'males_45age_avg_vote', 'males_45age_votes',
                                         'females_0age_avg_vote', 'females_0age_votes', 'females_18age_avg_vote', 'females_18age_votes', 'females_30age_avg_vote',
                                         'females_30age_votes', 'females_45age_avg_vote', 'females_45age_votes'])
IMDB_movies.head()
```

```python
# Adding positive votes, neutral votes, and negative votes columns
IMDB_movies['positive_votes'] = IMDB_movies['votes_10'] + IMDB_movies['votes_9'] + IMDB_movies['votes_8'] + IMDB_movies['votes_7']
IMDB_movies['neutral_votes'] = IMDB_movies['votes_6'] + IMDB_movies['votes_5'] + IMDB_movies['votes_4']
IMDB_movies['negative_votes'] = IMDB_movies['votes_3'] + IMDB_movies['votes_2'] + IMDB_movies['votes_1']
# Adding percentage of positive reviews column
IMDB_movies['percentage_of_positive_reviews'] = (IMDB_movies['positive_votes']/(IMDB_movies['positive_votes']+IMDB_movies['neutral_votes']+
  IMDB_movies['negative_votes'])).round(2)                                         
IMDB_movies.head()
```

```python
# Creating sentiment column from percentage of positive reviews (if more than 60% of reviews are positive then good movie)
IMDB_movies['overall_eval'] = np.where(IMDB_movies['percentage_of_positive_reviews']>=0.60,'good', 'bad')
IMDB_movies.head()
```

```python
# Converting good and bad values into dummies
IMDB_movies['overall_eval']=IMDB_movies['overall_eval'].str.get_dummies()
```

```python
IMDB_movies['overall_eval'].head()
```

```python
# Setting good = 1 and bad = 0
IMDB_movies['overall_eval'] = IMDB_movies['overall_eval'].replace({0:1, 1:0})
IMDB_movies['overall_eval']
```

```python
# Splitting genre into separate columns
IMDB_movies_analysis[['genre_1','genre_2', 'genre_3']] = IMDB_movies_analysis['genre'].str.split(",", expand=True)
IMDB_movies_analysis
```

```python
# Splitting country into separate columns
IMDB_movies_analysis[['country_1','country_2','country_3','rest']] = IMDB_movies_analysis['country'].str.split(",", n=3, expand=True)
IMDB_movies_analysis = IMDB_movies_analysis.drop(columns=['rest'])
IMDB_movies_analysis
```

```python
# Exporting dataframe to csv file
IMDB_movies_analysis.to_csv('IMDB_updated.csv')
```

```python
# Creating new dataframe for currency conversions
currency_conversions = IMDB_movies_analysis[['title','budget']]
currency_conversions[['currency_symbol', 'currency_amount']] = currency_conversions['budget'].str.split(" ", expand=True)
currency_conversions.head()
```

```python
# Identifying the different currencies
currency_conversions.currency_symbol.unique()
```

```python
# Converting currencies into USD
for index, row in currency_conversions.iterrows():
    if row['currency_symbol'] == 'DEM':
        row['currency_amount'] = float(row['currency_amount'])*0.001740
    if row ['currency_symbol'] == 'JPY':
        row['currency_amount'] = float(row['currency_amount'])*0.0093
    if row ['currency_symbol'] == 'AUD':
        row['currency_amount'] = float(row['currency_amount'])*0.65
    if row ['currency_symbol'] == 'RUR':
        row['currency_amount'] = float(row['currency_amount'])*0.013
    if row ['currency_symbol'] == 'GBP':
        row['currency_amount'] = float(row['currency_amount'])*1.26
    if row ['currency_symbol'] == 'CAD':
        row['currency_amount'] = float(row['currency_amount'])*0.72
    if row ['currency_symbol'] == 'FRF':
        row['currency_amount'] = float(row['currency_amount'])*0.17
    if row ['currency_symbol'] == 'DKK':
        row['currency_amount'] = float(row['currency_amount'])*0.15
    if row ['currency_symbol'] == 'SEK':
        row['currency_amount'] = float(row['currency_amount'])*0.10
    if row ['currency_symbol'] == 'ATS':
        row['currency_amount'] = float(row['currency_amount'])*0.078793604651163
    if row ['currency_symbol'] == 'EUR':
        row['currency_amount'] = float(row['currency_amount'])*1.10
    if row ['currency_symbol'] == 'SGD':
        row['currency_amount'] = float(row['currency_amount'])*0.71
    if row ['currency_symbol'] == 'HUF':
        row['currency_amount'] = float(row['currency_amount'])*0.0031
    if row ['currency_symbol'] == 'CNY':
        row['currency_amount'] = float(row['currency_amount'])*0.14
    if row ['currency_symbol'] == 'INR':
        row['currency_amount'] = float(row['currency_amount'])*0.013
    if row ['currency_symbol'] == 'KRW':
        row['currency_amount'] = float(row['currency_amount'])*0.00082
    if row ['currency_symbol'] == 'PLN':
        row['currency_amount'] = float(row['currency_amount'])*0.24
    if row ['currency_symbol'] == 'HKD':
        row['currency_amount'] = float(row['currency_amount'])*0.13
    if row ['currency_symbol'] == 'ARS':
        row['currency_amount'] = float(row['currency_amount'])*0.015
    if row ['currency_symbol'] == 'ILS':
        row['currency_amount'] = float(row['currency_amount'])*0.29
    if row ['currency_symbol'] == 'MXN':
        row['currency_amount'] = float(row['currency_amount'])*0.041
    if row ['currency_symbol'] == '$':
        row['currency_amount'] = int(row['currency_amount'])

currency_conversions.head()
```

```python
# Export dataframe to csv file
currency_conversions.to_csv('currencyConversions.csv')
```
